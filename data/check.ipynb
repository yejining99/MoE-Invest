```python
import pandas as pd

nasdaq100_bs = pd.read_csv("nasdaq100_bs.csv").drop_duplicates(subset=['COMPANYID', 'QUARTER', 'DATAITEMNAME'], keep='last')
nasdaq100_cf = pd.read_csv("nasdaq100_cf.csv").drop_duplicates(subset=['COMPANYID', 'QUARTER', 'DATAITEMNAME'], keep='last')
nasdaq100_is = pd.read_csv("nasdaq100_is.csv").drop_duplicates(subset=['COMPANYID', 'QUARTER', 'DATAITEMNAME'], keep='last')
nasdaq100_members = pd.read_csv("nasdaq100_members.csv")

nasdaq100_bs = nasdaq100_bs.pivot(index=['COMPANYID','QUARTER'], columns="DATAITEMNAME", values="DATAITEMVALUE")
nasdaq100_cf = nasdaq100_cf.pivot(index=['COMPANYID','QUARTER'], columns="DATAITEMNAME", values="DATAITEMVALUE")
nasdaq100_is = nasdaq100_is.pivot(index=['COMPANYID','QUARTER'], columns="DATAITEMNAME", values="DATAITEMVALUE")

# 전체 붙은 테이블
nasdaq100_all = pd.merge(nasdaq100_bs, nasdaq100_cf, on=['COMPANYID','QUARTER'], how='left')
nasdaq100_all = pd.merge(nasdaq100_all, nasdaq100_is, on=['COMPANYID','QUARTER'], how='left').reset_index()
#  COMPANYID랑 TICKERSYMBOL랑 붙이기
nasdaq100_all = pd.merge(nasdaq100_members[['COMPANYID', 'TICKERSYMBOL']], nasdaq100_all, on='COMPANYID', how='left')
```

```python
nasdaq100_all.to_csv("nasdaq100_bs_cf_is.csv", index=False)
```
